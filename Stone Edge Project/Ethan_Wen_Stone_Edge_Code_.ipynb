## Importing stuff, loading in links, functions... that kinda thing!

```python
# importing my funny little libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# loading the github csvs (thank you kayla)
rAndG_link = "https://raw.githubusercontent.com/kayla-owens/stars-summer-class/main/M_15_rg_crossmatch.csv"
rAndI_link = "https://raw.githubusercontent.com/kayla-owens/stars-summer-class/main/M_15_ri_crossmatch.csv"

# declaring my silly little functions
# im so quirky using a function
def convMag(list):
  for i in range(len(list)):
    list[i] = -2.5* np.log10(list[i])
```

## Code for Green (R) and Red (I) bands:

```python
# making the new variables to hold the data
rAndI = pd.read_csv(rAndI_link)

# pulling the separation data
rAndI_Sep = rAndI['Separation']

# pulling the flux data
rAndI_f1 = rAndI['flux_1']
rAndI_f2 = rAndI['flux_2']

# making magnitude arrays
rAndI_m1 = np.array(rAndI_f1)
rAndI_m2 = np.array(rAndI_f2)

# using the convMag function
# for the magnitudes, i chose to mirror the fluxes into the magnitude arrays, then use a function to convert the fluxes to the magnitudes :P
convMag(rAndI_m1)
convMag(rAndI_m2)

# setting up my separation boolean
rAndI_sepMatch = (4 < rAndI_Sep) & ( rAndI_Sep < 5)

# now filtering for for the values that fit
rAndI_m1_filtered = rAndI_m1[rAndI_sepMatch]
rAndI_m2_filtered = rAndI_m2[rAndI_sepMatch]

# histogram magic :sparkles:
plt.hist(rAndI_Sep, range=[4, 5])
plt.xlabel("Separation (pc)")
plt.ylabel("Number of Stars")
plt.title("Separation for M15, Green (R) and Red (I) Bands")
plt.show()

# CMD :sparkle:
fig, ax = plt.subplots()
ax.scatter((rAndI_m1_filtered - rAndI_m2_filtered), rAndI_m1_filtered, color='black', s=1)
ax.invert_yaxis()
ax.set_xlabel("Green (R band) - Red (I band) [Mag]")
ax.set_ylabel("Green (R band) [Mag]")
ax.set_title("CMD for Green (R) and Red (I) Bands")
```

## Code for Green (R) and Blue(G) bands:

```python
# making the new array to hold the data
rAndG = pd.read_csv(rAndG_link)

# pulling the separation data
rAndG_Sep = rAndG['Separation']

# pulling the flux data for rAndG
rAndG_f1 = rAndG['flux_1']
rAndG_f2 = rAndG['flux_2']

# making magnitude arrays
rAndG_m1 = np.array(rAndG_f1)
rAndG_m2 = np.array(rAndG_f2)

# converting to actual magnitudes instead of mrroring fluxes LMFAO i did not need a function
convMag(rAndG_m1)
convMag(rAndG_m2)

# separation boolean
rAndG_sepMatch = (5.95 < rAndG_Sep) & (rAndG_Sep < 6.5)

# now filtering for values that fit
rAndG_m1_filtered = rAndG_m1[rAndG_sepMatch]
rAndG_m2_filtered = rAndG_m2[rAndG_sepMatch]

# histogram magic :sparkles:
plt.hist(rAndG_Sep, range=[5.95, 6.5])
plt.xlabel("Separation (pc)")
plt.ylabel("Number of Stars")
plt.title("Separation for M15, Green (R) and Blue (G) Bands")
plt.show()

# CMD magic :interrobang:
fig, ax = plt.subplots()
ax.scatter((rAndG_m2_filtered - rAndG_m1_filtered), rAndG_m2_filtered, color='black', s=1)
ax.invert_yaxis()
ax.set_xlabel("Blue (G band) - Green (R band) [Mag]")
ax.set_ylabel("Blue (G band) [Mag]")
ax.set_title("CMD for Green (R) and Blue (G) Bands")
plt.show()
```

## Plotting RA and DEC for Green (R) and Red (I):

```python
# picking out the ra  and dec columns
# (I chose to use ra_1 and dec_1, i tried plotting both and it seems that they are the same so idk)
rAndI_ra1 = rAndI['ra_1']
rAndI_dec1 = rAndI['dec_1']

# plots!!!
plt.scatter(rAndI_ra1, rAndI_dec1, color='black', s=2)
# roblos :roblos:
plt.xlabel("RA (Degrees)")
plt.ylabel("DEC (Degrees)")
plt.title("Right Ascension and Declination, Green (R) and Red (I) bands")
plt.show()
```
