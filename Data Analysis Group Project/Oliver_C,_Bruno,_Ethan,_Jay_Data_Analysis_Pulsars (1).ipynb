# Data Manipulation

```python
# import our silly little imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# data
atnf_database = "https://raw.githubusercontent.com/wen-ethan/astr-work/main/Data%20Analysis%20Group%20Project/ATNF%20Database.csv"

# read the csv
pulsars = pd.read_csv(atnf_database)

#finding stuff
# print(pulsars.columns)
# print(pulsars.shape)

# I APOLOGIZE FOR THE HORRIBLE CODE >< I JUST WANTED TO GET IT DONE QUICK I WAS TOO LAZY TO FIND A BETTER WAY TO DO IT

# setting variables
age = pulsars['AGE']
dm = pulsars['DM']
epoch = pulsars['T0']
freq = pulsars['F0']
dist = pulsars['DIST']
lum = pulsars['R_LUM']
edot = pulsars['EDOT']
rm = pulsars['RM']


# creating the filtered_1 arrays
age_filtered = np.array([])
dm_filtered = np.array([])
epoch_filtered = np.array([])
freq_filtered = np.array([])
dist_filtered = np.array([])
lum_filtered = np.array([])
edot_filtered = np.array([])
rm_filtered = np.array([])

# an absurdly long for-loop to make sure there is no missing data
for i in range(len(age)):
      if not(age[i] == "*"):
        if not(dm[i] == "*"):
          if (not(epoch[i] == "*")):
            if(not(freq[i] == "*")):
              if (not(dist[i] == "*")):
                if(not(lum[i] == "*")):
                  if (not(edot[i] == "*")):
                    if(not(rm[i] == "*")):
                      age_filtered = np.append(age_filtered, age[i])
                      dm_filtered = np.append(dm_filtered, dm[i])
                      epoch_filtered = np.append(epoch_filtered, epoch[i])
                      freq_filtered = np.append(freq_filtered, freq[i])
                      dist_filtered = np.append(dist_filtered, dist[i])
                      lum_filtered = np.append(lum_filtered, lum[i])
                      edot_filtered = np.append(edot_filtered, edot[i])
                      rm_filtered = np.append(rm_filtered, rm[i])


#  coverting to floats (aka real numbers)
age_f2 = age_filtered.astype(np.float64)
dm_f2 = dm_filtered.astype(np.float64)
epoch_f2 = epoch_filtered.astype(np.float64)
freq_f2 = freq_filtered.astype(np.float64)
dist_f2 = dist_filtered.astype(np.float64)
lum_f2 = lum_filtered.astype(np.float64)
edot_f2 = edot_filtered.astype(np.float64)
rm_f2 = rm_filtered.astype(np.float64)

# removing bad datapoints
age_f2 = np.delete(age_f2, 4)
dm_f2 = np.delete(dm_f2, 4)
epoch_f2 = np.delete(epoch_f2, 4)
freq_f2 = np.delete(freq_f2, 4)
dist_f2 = np.delete(dist_f2, 4)
lum_f2 = np.delete(lum_f2, 4)
edot_f2 = np.delete(edot_f2, 4)
rm_f2 = np.delete(rm_f2, 4)

age_f2 = np.delete(age_f2, 0)
dm_f2 = np.delete(dm_f2, 0)
epoch_f2 = np.delete(epoch_f2, 0)
freq_f2 = np.delete(freq_f2, 0)
dist_f2 = np.delete(dist_f2, 0)
lum_f2 = np.delete(lum_f2, 0)
edot_f2 = np.delete(edot_f2, 0)
rm_f2 = np.delete(rm_f2, 0)
```
